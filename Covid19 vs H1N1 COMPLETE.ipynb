```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/2009_flu_pandemic_in_the_United_States'

# response = requests.get(url)
# soup = BeautifulSoup(url, 'html.parser')
```

```python
# table_a = soup.find_all("table", class_='navbar')
tables = pd.read_html(url)
tables
```

```python
df = tables[2]

df.head(58)
```

```python
df = df.rename(columns={"State or territory":"states", 
                        "State-reported confirmed cases[1]": "h1n1_confirmed_cases", 
                        "State-reported hospital- izations[1]": "h1n1_hospitalization", 
                        "Confirmed deaths": "h1n1_confirmed_deaths"})
del df["h1n1_hospitalization"]
```

```python
df= df.dropna()
df
```

```python
df["h1n1_confirmed_cases"] = df["h1n1_confirmed_cases"].apply(lambda x:x.split('/NLU')[0])
df["h1n1_confirmed_cases"] = df["h1n1_confirmed_cases"].apply(lambda x:x.split('[')[0])
df["h1n1_confirmed_deaths"] = df["h1n1_confirmed_deaths"].apply(lambda x:x.split('[')[0])
df["h1n1_confirmed_deaths"] = df["h1n1_confirmed_deaths"].apply(lambda x:x.split('wave 2')[0])
```

```python
df['h1n1_confirmed_cases'] = df['h1n1_confirmed_cases'].str.replace(',', '')

df['h1n1_confirmed_cases']
```

```python
df
```

```python
df.to_csv(r'Resources\H1N1.csv', index=False)
```

```python
url_2 = 'https://www.worldometers.info/coronavirus/country/us/'
response=requests.get(url_2)
tables_2 = pd.read_html(response.text)
df2= tables_2[0]
```

```python
df2 = df2.rename(columns={"USAState":"states", 
                              "TotalCases": "covid19_confirmed_cases", 
                              "NewCases": "covid19_new_case",
                              "TotalDeaths": "covid19_confirmed_deaths",
                              "NewDeaths": "covid19_new_deaths",
                              "ActiveCases": "covid19_active_cases"})
df2
```

```python
Corona_df = df2[["states", "covid19_confirmed_cases", "covid19_confirmed_deaths"]]
```

```python
Corona_df.to_csv(r'Resources\Covid19.csv', index=False)
```
